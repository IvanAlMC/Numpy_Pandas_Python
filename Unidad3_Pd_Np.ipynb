## **Numpy Arange**

numpy.arange([start, ]stop, [step, ]dtype=None, *, like=None)

Devuelve valores espaciados uniformemente dentro de un intervalo determinado.

Arange se puede llamar con un número variable de argumentos posicionales:

- arange(stop): Los valores se generan dentro del intervalo medio abierto [0, stop) (en otras palabras, el
intervalo que incluye el inicio pero excluye el stop).

- arange(start, stop): Los valores se generan dentro del intervalo medio abierto [start, stop).

- arange(start, stop, step) Los valores se generan dentro del intervalo medio abierto [start, stop), con un espacio entre valores dado por paso.

Fuente:https://numpy.org/doc/stable/reference/generated/numpy.arange.html

```python
#crear un arreglo de números que comienzan en 0 (incluido) y terminan en 5 (excluido), avanzando en incrementos de 0.5
print(np.arange(0, 5, 0.5))
```

```python
#crear un arreglo de nueros que comienza en -20, avanza en incrementos de 2 y termina en 20 (excluido)
print(np.arange(-20, 20, 2, dtype=int))
```

```python
#crear un arreglo con números generados que comienza en -20, avanza en incrementos de 0.5 y termina en 20 (excluido)
print(np.arange(-20, 20, 0.5, dtype=float))
```

## **Filtrado de elementos de un array**

Una característica muy útil de los arrays es que es muy fácil obtener otro array con los elementos que cumplen una condición. a[condicion] : Devuelve un arreglo de una dimensión con los elementos del array que cumplen la condición.


```python
a = np.array([[1, 2, 3], [4, 5, 6]])
print(a)
```

```python
# filtrando los elementos de la matriz a que son pares
print(a[(a % 2 == 0)])
```

```python
#filtrando los elementos pares de la matriz a que son mayores que 2.
print(a[(a % 2 == 0) & (a > 2)])
```

```python
#filtrando todos los elementos de la matriz a que son mayores o iguales a 1
print(a[(a>=1)])
```

## **Slicing and memory areas Arrays**

```python
# ----------------------------- Basic Indexing and Slicing
arr = np.arange(10)
print(arr) #Se imprime el arreglo creado anteriomente con np.arange
print(arr[5]) #Se obtiene el valor de la posicion 5 del arreglo
print(arr[5:8]) #Se obtiene los valores entre la posicion 5 y 7 del arreglo
arr[5:8] = 12 #Se cambian los valores de la posicion 5 a 7 por 12
print(arr)
```

```python
arr_slice = arr[5:8] # apuntan a la misma area de memoria ? ?
print(arr_slice) #Apunta a los valores de las posiciones 5 hasta la 7 del anterior arreglo
print(arr_slice.ndim) #Se imprime su dimension
print(arr_slice.shape) #Se obtiene la tupla
print(arr_slice.dtype) #Su tipo
arr_slice[2]=30 #se modifica el valor en la posicion 2 por 30 por el arreglo que apunta a los valores de la posicion 5 a 7
print(arr) #Se imprime el arreglo principal
print(arr_slice) #Se imprime el pedazo de arreglo dentro del arreglo principal
```

```python
#Se hace el cambio de pedazo del arreglo de la posicion 5 a 7 en la posicion 1 de dicho sub arreglo a 12345
arr_slice[1] = 12345
print(arr)
print(arr_slice)
```

```python
#Se hace el cambio de pedazo del arreglo de la posicion 5 a 7 en todas las posiciones de dicho sub arreglo a 64
arr_slice[:] = 64
print(arr)
print(arr_slice)
```

```python
#Se hace una copia que no afecta a el arreglo principal o al subarreglo
arr2=arr[5:8].copy() # es una copia y queda en otra area de memoria
print(arr2)
arr2[:]=3
print(arr)
print(arr_slice)
print(arr2)
```

**bincount():** La función bincount() cuenta la cantidad de ocurrencias de valores enteros no negativos en un arreglo. Es útil para el análisis de datos discretos.

Ejemplo:

```python
import numpy as np
arr = np.array([1, 1, 2, 2, 3, 3, 3, 3])
counts = np.bincount(arr)
print(counts)  # [0 1 2 3]
```

**bincount().argmax():** La función argmax() aplicada a la salida de bincount() te devuelve el valor más frecuente (la moda) en un arreglo de enteros no negativos.

Ejemplo:

```python
import numpy as np
arr = np.array([1, 2, 2, 3, 3, 3])
counts = np.bincount(arr)
mode = np.argmax(counts)
print(mode)  # 3
```

**hstack():** La función hstack() se utiliza para apilar varios arreglos horizontalmente (a lo largo de las columnas).

Ejemplo:

```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
stacked = np.hstack((arr1, arr2))
print(stacked)  # [1 2 3 4 5 6]
```

**vstack():** La función vstack() se utiliza para apilar varios arreglos verticalmente (a lo largo de las filas).

Ejemplo:

```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
stacked = np.vstack((arr1, arr2))
print(stacked)
# [[1 2 3]
#  [4 5 6]]
```

**dstack():** La función dstack() se utiliza para apilar arreglos a lo largo del tercer eje (profundidad).

Ejemplo:

```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
stacked = np.dstack((arr1, arr2))
print(stacked)
# [[[1 4]
#  [2 5]
#  [3 6]]
```

## **Ejercicio 4**

Reversar un arreglo de numpy 1D, Sin usar funciones específicas de numpy:
Arreglo inicial [1 2 3 4 5 6]
Salida [6 5 4 3 2 1]

```python
import numpy as np

# Arreglo inicial
arr = np.array([1, 2, 3, 4, 5, 6])

# Longitud del arreglo
n = len(arr)

# Crear un nuevo arreglo para almacenar el resultado
reversed_arr = np.empty_like(arr)

# Revertir el arreglo manualmente
for i in range(n):
    reversed_arr[i] = arr[n - i - 1]

# Imprimir el arreglo invertido
print(reversed_arr)
```

## 🌵DataFrame

Un DataFrame es una estructura de datos que almacena la información como una tabla ordenada por filas y columnas. Cada fila representa un objeto y cada columna la información correspondiente a una característica de los objetos.

Un DataFrame también posee índices por cada fila, que pueden ser dados o generados automáticamente. Cada columna del DataFrame es una [serie], donde el valor del índice corresponde con los valores de índice que tiene el DataFrame.

Por medio de un diccionario vamos a crear un dataframe, donde las llaves son los nombres de las columnas y los valores son las características.

Por ejemplo, para hacer un DataFrame con el equipo de fútbol anterior, pero agregando las series de estatura y peso, se hará de la siguiente manera:


```python
import pandas as pd
dict_caracteristicas = {'apellido':['Ospina', 'Zapata', 'Falcao', 'Cuadrado', 'Rodriguez'],
'altura':[183.0,187.0,177.0,179.0,180.0],
'peso':[80.0,82.0,72.0,72.0,75.0]}
seleccionColombia = pd.DataFrame(dict_caracteristicas,index=[1, 2, 9, 11, 10])
```

2. Consultar sobre caracteristicas, gestión, carga y creación de archivos en formato HDF5. Proponer y realizar un ejercicio completo.


El formato HDF5 (Hierarchical Data Format version 5) es un formato de archivo binario ampliamente utilizado en la ciencia de datos y la ingeniería para almacenar y gestionar grandes conjuntos de datos de manera eficiente. HDF5 es especialmente adecuado para datos multidimensionales y jerárquicos. A continuación, te mostraré cómo crear, gestionar y cargar archivos HDF5 utilizando Python y la biblioteca pandas.

Primero, asegúrate de tener instalada la biblioteca pandas y h5py (una biblioteca comúnmente utilizada para trabajar con archivos HDF5) en tu entorno de Python.

 Un ejercicio completo que abarca la creación, gestión y carga de archivos HDF5 con pandas:

```python
import pandas as pd
import h5py

# Crear un DataFrame de ejemplo
data = {
    'Nombre': ['Alice', 'Bob', 'Carol'],
    'Edad': [25, 30, 35],
    'Salario': [50000, 60000, 75000]
}

df = pd.DataFrame(data)

# Crear un archivo HDF5 y guardar el DataFrame en él
with pd.HDFStore('ejemplo.h5', mode='w') as store:
    store['mi_dataframe'] = df

print("Archivo HDF5 creado y DataFrame guardado.")
```

En este ejemplo, hemos creado un DataFrame llamado df y lo hemos guardado en un archivo HDF5 llamado "ejemplo.h5" utilizando pd.HDFStore.

**Gestión de archivos HDF5:**

Puedes gestionar archivos HDF5 para agregar, actualizar o eliminar datos. A continuación, se dan ejemplos de cómo hacerlo:

Agregar datos a un archivo HDF5:

```python
import pandas as pd
import numpy as np

# Crear un DataFrame de ejemplo
data = {
    'A': np.random.randn(1000),
    'B': np.random.randint(0, 2, size=1000),
    'C': np.random.choice(['X', 'Y', 'Z'], size=1000)
}
df = pd.DataFrame(data)

# Escribir el DataFrame en un archivo HDF5
df.to_hdf('ejemplo.h5', key='mi_data', mode='w')

# Cargar el DataFrame desde el archivo HDF5
df_cargado = pd.read_hdf('ejemplo.h5', key='mi_data')

# Visualizar el DataFrame cargado
print(df_cargado.head())

# Realizar una consulta en el DataFrame cargado
consulta = df_cargado[df_cargado['B'] == 1]
print(consulta.head())
```

Pandas es una excelente biblioteca para trabajar con datos tabulares, pero generalmente se utiliza para cargar y manipular datos en formatos como CSV, Excel, SQL y más. No es común utilizarlo para cargar, gestionar o crear archivos HTML, ya que los archivos HTML se utilizan principalmente para crear páginas web y no para almacenar datos tabulares. Sin embargo, puedo proporcionarte información sobre cómo trabajar con HTML en Python y Pandas.

Si deseas crear, gestionar o cargar archivos HTML, necesitarás usar otras bibliotecas específicas para HTML, como Beautiful Soup para el análisis HTML o una biblioteca para la creación de páginas web.

Por ejemplo, si deseas crear un archivo HTML simple usando Python

```python
# Importar la biblioteca para crear un archivo HTML
from bs4 import BeautifulSoup

# Crear un objeto BeautifulSoup
html = BeautifulSoup("<html><body><h1>Mi Página HTML</h1><p>Este es un párrafo.</p></body></html>", "html.parser")

# Guardar el archivo HTML en disco
with open("mi_pagina.html", "w") as file:
    file.write(html.prettify())
```

Este código crea un archivo HTML simple con un título y un párrafo. Puedes modificar el contenido HTML según tus necesidades.

Si deseas cargar y analizar datos de un archivo HTML, Beautiful Soup es una excelente opción para analizar la estructura del documento HTML y extraer información. También puedes utilizar Pandas para cargar datos de tablas HTML en un DataFrame si el HTML tiene tablas.

```python
import pandas as pd
from bs4 import BeautifulSoup

# Leer el archivo HTML
with open("mi_tabla.html", "r") as file:
    html_content = file.read()

# Crear un objeto BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")

# Encontrar la tabla en el HTML
table = soup.find("table")

# Cargar la tabla en un DataFrame de Pandas
df = pd.read_html(str(table))[0]

# Ahora puedes trabajar con el DataFrame df
print(df)
```

4. Consultar como crear un archivo PDF a partir de un DataFrame. Proponer y realizar un ejercicio completo

Para crear un archivo PDF a partir de un DataFrame en Python, puedes usar la biblioteca pandas para manejar los datos del DataFrame y la biblioteca reportlab para generar el PDF.

```python
pip install reportlab
```

```python
import pandas as pd
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Table, TableStyle
from reportlab.lib import colors

# Crear un DataFrame de ejemplo
data = {
    'Nombre': ['Alice', 'Bob', 'Carol', 'David', 'Emily'],
    'Edad': [28, 32, 45, 22, 38],
    'Salario': [50000, 60000, 75000, 42000, 68000]
}

df = pd.DataFrame(data)

# Crear un archivo PDF
pdf_filename = "dataframe_to_pdf.pdf"
pdf = SimpleDocTemplate(pdf_filename, pagesize=letter)

# Convertir el DataFrame a una lista de listas
data = [df.columns.tolist()] + df.values.tolist()

# Crear una tabla a partir de los datos
table = Table(data)

# Establecer el estilo de la tabla
style = TableStyle([
    ('BACKGROUND', (0, 0), (-1, 0), colors.grey),  # Encabezado
    ('TEXTCOLOR', (0, 0), (-1, 0), colors.whitesmoke),  # Color del texto en el encabezado
    ('ALIGN', (0, 0), (-1, -1), 'CENTER'),  # Alineación al centro
    ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),  # Fuente en el encabezado
    ('BOTTOMPADDING', (0, 0), (-1, 0), 12),  # Espacio inferior en el encabezado
    ('BACKGROUND', (0, 1), (-1, -1), colors.beige),  # Color de fondo para el contenido
    ('GRID', (0, 0), (-1, -1), 1, colors.black)  # Líneas de la cuadrícula
])

table.setStyle(style)

# Crear el PDF
story = []
story.append(table)
pdf.build(story)

print(f"Se ha creado el archivo PDF: {pdf_filename}")
```

5. Crear un archivo excel a partir de un dataframe. El ejemplo debe contemplar el almacenamiento en diferentes hojas de un libro de excel. El nombre de cada hoja corresponderá a valores distintos de la serie Team y a su vez en cada hoja se almacenará los empleados que pertenecen a cada uno de los equipos.

```python
pip install XlsxWriter
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'Team': ['EquipoA', 'EquipoB', 'EquipoA', 'EquipoC', 'EquipoB'],
    'Employee': ['Empleado1', 'Empleado2', 'Empleado3', 'Empleado4', 'Empleado5']
}

df_emp = pd.DataFrame(data)

# Crear un archivo Excel y almacenar diferentes hojas
with pd.ExcelWriter('empleados_por_equipo.xlsx', engine='xlsxwriter') as writer:
    # Obtener valores únicos en la columna 'Team'
    unique_teams = df_emp['Team'].unique()

    # Iterar a través de los valores únicos y crear una hoja por equipo
    for team in unique_teams:
        # Filtrar el DataFrame para obtener empleados de un equipo en particular
        team_df = df_emp[df_emp['Team'] == team]

        # Escribir el DataFrame en la hoja correspondiente
        team_df.to_excel(writer, sheet_name=team, index=False)

print("Archivo Excel creado exitosamente.")
```
